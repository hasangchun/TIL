```python
# merge
df3 = pd.DataFrame({'cus_no': [1001,1002,1003,1004],
                   'name': ['choi','kim','park','hur']})
df3
```

```python
df4 = pd.DataFrame({'cus_no': [1001,1001,1003,1006],
                   'amnt': [10000,20000,15000,5000]})
df4
```

```python
pd.concat([df3, df4])
```

```python
pd.concat([df3, df4], axis = 1)
```

```python
pd.merge(df3, df4)  # join only the common columns (inner join)
```

```python
pd.merge(df3, df4, how='outer')  #  outer join 
```
