# Linear Regression (선형회귀)
### Ex1: Linear regression with two variables (y = wx + b)

```python
# ex1: linear regression with two variables (y = wx + b)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

n = 100
x = np.random.randn(n)                # batch size
y = x*20 + 10                         # w=20, b=10
y = y + np.random.randn(n) * 10       # add noise

# x,y
#### just to see the data
plt.figure(figsize=(4,4)) # plotting a circle
plt.scatter(x,y)
##################\

w=np.random.randn()   # randn = normal distribution: meqn=0, signma=1, rand= (0,1)
b=np.random.randn()

lr = 0.2                                 # learning rate
n_epoch = 200

for epoch in range(n_epoch):
    y_pred = w*x + b
    error = ((y_pred - y)**2).mean()     # mean square error

    w = w - lr* ((y_pred - y)*x).mean()
    b = b - lr* (y_pred - y).mean()
    if epoch %10 == 0:
        print('epoch=', epoch, 'error=', error, 'w=', w, 'b=', b)
        
print('---------------------------')
print('epoch=', epoch, 'error=', error, 'w=', w, 'b=', b)

xx = np.linspace(-5,5,100) 
yy = w * xx + b
plt.plot(xx,yy,c='k') 
```

### Ex2: Training two features (x1, x2)  ( y = w1*x1 + w2*x2 + b)

```python
# ex 2 : training two parameters w1, w2 and b (y = w1*x1 + w2*x2 + b)

import numpy as np
import pandas as pd

n=100
x1 = np.random.randn(n)             # randn=normal distribution in (-1,1), rand=(0,1)
x2 = np.random.randn(n)

y = x1*30 + x2*40 + 50
y = y + np.random.randn(n)*20      # add noise

w1 = np.random.rand()               # initial guess
w2 = np.random.rand()
b = np.random.rand()

lr = 0.2                            # learning rate
n_epoch = 200                      # no of epoch
for epoch in range(n_epoch):
    y_pred = w1*x1 + w2*x2 + b
    error = ((y_pred - y)**2).mean()

    w1 = w1 - lr* ((y_pred - y)*x1).mean()
    w2 = w2 - lr* ((y_pred - y)*x2).mean()
    b = b - lr* (y_pred - y).mean()
    if epoch %10 == 0:
        print('epoch=', epoch, 'error=', error, 'w1=', w1, 'w2=', w2, 'b=', b)
        
print('---------------------------')
print('epoch=', epoch, 'error=', error, 'w1=', w1, 'w2=', w2, 'b=', b)
```

# Classification (선형분류)
### Ex1: Linear classification with two variables
- data generated by myself
- uses MSE as loss function (error = (sigmoid(x) - y)**2)

```python
### for classification
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
N = 200
x1_p = np.random.randn(int(N/2))
x2_p = -1.7 * x1_p + 0.8 + np.random.rand(n) * 1.7
x1_n = np.random.randn(int(N/2))
x2_n = -1.7 * x1_n + 0.8 - np.random.rand(n) * 2.4

# plt.scatter(x1_p, x2_p)
# plt.scatter(x1_n, x2_n)
# xx = np.linspace(-4, 4,100)
# plt.plot(xx, 0.7*xx + 1.3)
# plt.plot()
# plt.show()

X1 = np.concatenate((x1_p, x1_n))
X2 = np.concatenate((x2_p, x2_n))

X = np.concatenate((X1.reshape(-1,1), X2.reshape(-1,1)), axis=1)

bias_factor = np.ones(X.shape[0]).reshape(-1,1)
X = np.concatenate((X, bias_factor), axis=1)
X.shape

y1=np.ones(len(x1_p))
y0=np.zeros(len(x1_n))
y = np.hstack((y1,y0))

W = np.random.randn(X.shape[1])   # initial W value

def sigmoid_activation(x):
    return 1.0 / (1 + np.exp(-x))

lossHistory = []
epochs = 100
alpha = 0.1

for epoch in np.arange(epochs):
    preds = sigmoid_activation(X.dot(W))
    error = preds - y
    loss = (error**2).sum()    # mean squares error
    lossHistory.append(loss)
    #  print("[INFO] epoch #{}, loss={:.7f}".format(epoch + 1, loss))
    
    gradient = X.T.dot(error) / X.shape[0]
    W = W - alpha * gradient

w1, w2, b = W[0], W[1], W[2]
x1, x2 = X[:,0], X[:,1]
print(w1, w2, b)
accuracy = ((sigmoid_activation(w1*x1 + w2*x2 + b) > 0.5) == y).sum()/N
print(accuracy)

plt.scatter(X[:,0], X[:,1], c=y)
xx = np.linspace(-3,6,100)
yy = -w1/w2 * xx -b/w2
plt.plot(xx, yy)
```

```python
# construct a figure that plots the loss over time
fig = plt.figure()
plt.ylim([0.,90])
plt.plot(np.arange(0, epochs), lossHistory)
fig.suptitle("Training Loss")
plt.xlabel("Epoch #")
plt.ylabel("Loss")
plt.show()
```

# Ex4: Using Cross Entropy as loss function (*******************)
- Classification by Calculating parameters one by one
- using CrossEntropy  as loss function
- number of features = 2

```python
from sklearn.datasets import make_blobs

np.random.seed(7)
N = 300
(X, y) = make_blobs(n_samples=N, n_features=2, centers=2, cluster_std=1.5, random_state=17)
x1, x2 = X[:,0], X[:,1]

w1 = np.random.randn()   # initial W value
w2 = np.random.randn()   # initial W value
b = np.random.randn()   # initial W value

def sigmoid_activation(x):
    return 1.0 / (1 + np.exp(-x))

lossHistory = []
epochs = 300
alpha = 0.01

for epoch in np.arange(epochs):
    preds = sigmoid_activation(w1*x1 + w2*x2 + b)       # prediction
    # error = preds - y
    loss = -((y*np.log(preds) + (1-y)*np.log(1-preds))).mean()  # loss = cross entropy
    lossHistory.append(loss)
    
    dloss_dz = preds - y
    w1_deriv = dloss_dz * x1        # d(loss)/dw1 = d(loss)/dz * dz/dw1
    w2_deriv = dloss_dz * x2
    b_deriv = dloss_dz * 1
    
    w1 = w1 - (alpha * w1_deriv).mean()
    w2 = w2 - (alpha * w2_deriv).mean()
    b  = b  - (alpha * b_deriv).mean()

print(w1, w2, b)
accuracy = ((sigmoid_activation(w1*x1 + w2*x2 + b) > 0.5) == y).sum()/N
print(accuracy)

# plt.ylim([-2,15])
plt.scatter(X[:,0], X[:,1], c=y)
xx = np.linspace(-10,5,100)
yy = -w1/w2 * xx -b/w2
plt.plot(xx, yy)
plt.show()

# construct a figure that plots the loss over time
fig = plt.figure()
plt.plot(np.arange(0, epochs), lossHistory)
plt.ylim([0.,1.])
fig.suptitle("Training Loss")
plt.xlabel("Epoch #")
plt.ylabel("Loss")
plt.show()
```

# Cross entropy 연습
- 
- This is the most common setting for classification problems. Cross-entropy loss increases as the predicted probability diverges from the actual label.
- Notice that when actual label is 1 (y(i) = 1), second half of function disappears whereas in case actual label is 0 (y(i) = 0) first half is dropped off. 
- In short, we are just multiplying the log of the actual predicted probability for the ground truth class. An important aspect of this is that cross entropy loss penalizes heavily the predictions that are confident but wrong.

```python
import numpy as np

def cross_entropy(predictions, targets, epsilon=1e-12):
    """
    Computes cross entropy between targets (encoded as one-hot vectors)
    and predictions. 
    Input: predictions (N, k) ndarray
           targets (N, k) ndarray        
    Returns: scalar
    """
    predictions = np.clip(predictions, epsilon, 1. - epsilon)
    N = predictions.shape[0]
    ce = -np.sum(targets*np.log(predictions+1e-9))/N
    return ce

predictions = np.array([[0.25,0.25,0.25,0.25],    # multi-class?
                        [0.01,0.01,0.01,0.96]])
targets = np.array([[0,0,0,1],
                    [0,0,0,1]])
ans = 0.71355817782  #Correct answer
x = cross_entropy(predictions, targets)
print(np.isclose(x,ans))
```

```python
cross_entropy_loss = cross_entropy(predictions, targets)
print ("Cross entropy loss is: " + str(cross_entropy_loss))
```
